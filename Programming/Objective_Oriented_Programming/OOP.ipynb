```python
class Dog:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    def set_age(self, age):
        self.age = age

d = Dog("Tim", 34)
d2 = Dog("Bill", 12)
print(d.get_name())
print(d.get_age())

print(d2.get_name())
print(d2.get_age())

d.set_age(23)
print(d.get_age())



```

## When to Use self:

- Instance variables that need to be accessed across different methods
- Data that needs to persist throughout the object's lifetime
- Properties that define the object's state

## When NOT to Use self:

- Temporary calculation variables
- Loop counters
- Method-specific helper variables
- Any variable only needed within a single method


```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def get_grade(self):
        return self.grade
    
class Course:
    def __init__(self, name, max_students):
        self.name = name
        self.max_students = max_students
        self.students = []  

    def add_student(self, student): 
        if len(self.students) < self.max_students:
            self.students.append(student)
            return True
        return False
    
    def get_average_grade(self):
        value = 0
        for student in self.students:
            value += student.get_grade()
        return value / len(self.students)

s1 = Student("Tim", 19, 95)
s2 = Student("Bill", 19, 75)
s3 = Student("Jill", 19, 65)

course = Course("Science", 3)
course.add_student(s1)
course.add_student(s2)
course.add_student(s3)
print(course.get_average_grade())
print(course.students[0].name)
```

# Inheritance

Inheritance allows us to create a new class that is a modified version of an existing class. Think of it like a child inheriting traits from their parents, but being able to have their own unique characteristics too


```python
class Pet:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old")
        
    def speak(self):
        print("I don't know what to say")

# Child class (inherits from Pet)
class Cat(Pet):

    def __init__(self, name1, age1, color): 

        super().__init__(name1, age1) #super() allows us to use the Parent's __init__ to not repeat code
        # These values get passed to parent's __init__
        # Parent creates:
        #   self.name = name1
        #   self.age = age1
        self.color = color

    def show2(self):  # Override the show method
        print(f"I am {self.name} and I am {self.age} years old and I am {self.color}")

# Another child class
class Dog(Pet):
    pass
    
# Let's try it out
pet = Pet("Tim", 19)
cat = Cat("Bill", 7, "Brown")
dog = Dog("Jill", 4)

pet.show()  # Uses Pet's show method
cat.show()  # Cat inherited show method from Pet

pet.speak()  
cat.speak()  
dog.speak()  

print()

cat.show2()
```

# Polymorphism

The ability of different classes to be treated as instances of the same parent class, typically by overriding methods. It allows for a single interface to work with different types of objects.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def study(self):
        print("Studying general subjects")

class ScienceStudent(Student):
    def study(self):  # Same method name, different behavior
        print("Studying science experiments")

class ArtStudent(Student):
    def study(self):  # Same method name, different behavior
        print("Studying art history")

# Using polymorphism
students = [
    ScienceStudent("Tim", 19),
    ArtStudent("Bill", 20)
]

for student in students:
    student.study()  # Each student studies differently
```

# Encapsulation

Definition: The practice of restricting direct access to certain attributes and methods, typically by making them private or protected, and providing controlled access through methods (getters and setters).

```python
class Student:
    def __init__(self, name, age):
        self._name = name  # Protected attribute (single underscore)
        self.__age = age   # Private attribute (double underscore)
        self.__grades = []
    
    def add_grade(self, grade):
        if 0 <= grade <= 100:  # Validate before adding
            self.__grades.append(grade)
            return True
        return False
    
    def get_average(self):
        if not self.__grades:
            return 0
        return sum(self.__grades) / len(self.__grades)
    
    @property
    def name(self):  # Getter for name
        return self._name
    
    @name.setter
    def name(self, value):  # Setter for name
        if isinstance(value, str) and value.strip():
            self._name = value.strip()

# Usage
student = Student("Tim", 19)
student.add_grade(95)
student.add_grade(87)
print(student.get_average())
# print(student.__grades)  # This would raise an error - can't access private attribute
```

# Abstraction

Abstraction is about hiding complex implementation details and showing only the necessary features. Here's an example using abstract classes (it also enforces that every child class must use the abstract methods)

```python
from abc import ABC, abstractmethod

class Assessment(ABC):
    @abstractmethod
    def calculate_score(self):
        pass
    
    @abstractmethod
    def get_grade(self):
        pass

class Quiz(Assessment):
    def __init__(self, questions, answers):
        self.questions = questions
        self.answers = answers
    
    def calculate_score(self):
        # Complex implementation hidden from user
        correct = sum(q == a for q, a in zip(self.questions, self.answers))
        return (correct / len(self.questions)) * 100
    
    def get_grade(self):
        score = self.calculate_score()
        if score >= 90: return 'A'
        if score >= 80: return 'B'
        if score >= 70: return 'C'
        return 'F'

class Project(Assessment):
    def __init__(self, requirements_met, code_quality, documentation):
        self.requirements_met = requirements_met
        self.code_quality = code_quality
        self.documentation = documentation
    
    def calculate_score(self):
        # Different complex implementation
        return (self.requirements_met * 0.5 +
                self.code_quality * 0.3 +
                self.documentation * 0.2) * 100
    
    def get_grade(self):
        score = self.calculate_score()
        if score >= 90: return 'A'
        if score >= 80: return 'B'
        if score >= 70: return 'C'
        return 'F'

# Usage
quiz = Quiz(['Q1', 'Q2', 'Q3'], ['A1', 'A2', 'A3'])
project = Project(0.95, 0.85, 0.90)

assessments = [quiz, project]
for assessment in assessments:
    print(f"Grade: {assessment.get_grade()}")
```
