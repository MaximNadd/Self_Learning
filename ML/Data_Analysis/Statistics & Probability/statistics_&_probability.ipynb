# RANDOM VARIABLES

A random variable is a numerical value assigned to the outcome of a random experiment. Random variables are used in probability and statistics to quantify uncertainty.

There are two types of random variables:


Discrete Random Variables: These take on a countable number of distinct values (e.g., whole numbers like 0, 1, 2, 3).

Continuous Random Variables: These take on an infinite number of possible values within a range (e.g., any number in [0,1] or [−∞,∞]).

### Real-Life Examples
Discrete Random Variable:

X: Number of cars sold in a dealership in a day ( X= 0,1,2,…).

Continuous Random Variable:

Y: Time a customer spends shopping in a store (e.g.,  Y∈[0,∞)).


# PROBABILITY DISTRIBUTIONS

#### What is a Probability Distribution?

A probability distribution defines the probabilities of the possible outcomes of a random variable. The outcomes can be:

    Discrete: A finite or countably infinite number of outcomes.

    Continuous: An uncountably infinite number of outcomes over a range of values.

#### Types of Probability Distributions
A) Discrete Probability Distributions

Discrete distributions deal with random variables that have countable outcomes. For example, the outcome of rolling a die or counting the number of heads in coin flips.


Bernoulli Distribution: Describes a single trial with two possible outcomes: success or failure. It is parameterized by the probability of success 
p (where 
0≤p≤1).


    Example: Flipping a coin (Heads = success, Tails = failure).

Binomial Distribution: Describes the number of successes in a fixed number of independent Bernoulli trials. It is parameterized by the number of trials 
𝑛
n and the probability of success 
𝑝
p.

    Example: The number of heads in 10 coin flips.


Poisson Distribution: Describes the number of events that occur in a fixed interval of time or space, with a known constant rate 
λ (average rate of occurrence).

    Example: The number of cars passing through a toll booth per hour.

B) Continuous Probability Distributions

Continuous distributions deal with random variables that can take any value within a certain range (often over an interval). The probability of any exact value is zero, but the probability of the variable falling within a specific range is non-zero.


Normal Distribution (Gaussian Distribution): The most common distribution in statistics, often used to model real-world phenomena like height, weight, or test scores. It is symmetric around its mean 
μ and is defined by its mean and standard deviation 
𝜎.

    Example: The distribution of heights of people in a population.

Uniform Distribution: Every outcome in the range is equally likely. It is often used for modeling situations with equal likelihood of outcomes within a certain range.

    Example: The distribution of outcomes when a fair die is rolled.

Exponential Distribution: Models the time between events in a Poisson process. It is parameterized by a rate 
λ and is often used to model waiting times or lifetimes of objects.

    Example: The time between arrivals of customers at a service desk.



### Key Concepts

a) Probability Mass Function (PMF)
##### For discrete distributions, the PMF gives the probability of each possible outcome. For a discrete random variable 𝑋, the PMF is denoted as 𝑃(𝑋=𝑥), and the sum of all probabilities in the distribution must equal 1


b) Probability Density Function (PDF)

The Probability Density Function (PDF) describes the likelihood of a continuous random variable taking a specific value, by representing the density of the probability at that point. The area under the curve of the PDF over a given range represents the probability that the variable falls within that range. The probability at a single point is always zero for continuous distributions.



c) Cumulative Distribution Function (CDF)

A Cumulative Distribution Function (CDF) tells you the probability of a random variable being less than or equal to a specific value. The CDF starts at 0 and grows to 1 as the value increases. For continuous variables, it’s the area under the PDF curve from 
−∞ to the given value. For discrete variables, it’s the sum of the probabilities of all outcomes less than or equal to the value.

The Cumulative Distribution Function (CDF) gives the proportion (or percentage) of people, items, or events that are at or below a certain value. It essentially accumulates probabilities up to that point.

example: F(180)=0.691:
This means 69.1% of adult males are 180 cm or shorter.






```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


# DISCRETE DISTRIBUTIONS

# 1. Bernoulli Distribution
p = 0.6  # probability of success
x_bernoulli = [0, 1]
pmf_bernoulli = [stats.bernoulli.pmf(k, p) for k in x_bernoulli]

plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.bar(x_bernoulli, pmf_bernoulli, color='skyblue', alpha=0.7)
plt.title("Bernoulli Distribution PMF")
plt.xlabel("Outcome (0 or 1)")
plt.ylabel("Probability")

# 2. Binomial Distribution
n = 10  # number of trials
p_binomial = 0.5  # probability of success
x_binomial = np.arange(0, n+1)
pmf_binomial = stats.binom.pmf(x_binomial, n, p_binomial)

plt.subplot(132)
plt.bar(x_binomial, pmf_binomial, color='lightgreen', alpha=0.7)
plt.title("Binomial Distribution PMF")
plt.xlabel("Number of successes")
plt.ylabel("Probability")

# 3. Poisson Distribution
lambda_poisson = 3  # average number of events
x_poisson = np.arange(0, 11)
pmf_poisson = stats.poisson.pmf(x_poisson, lambda_poisson)

plt.subplot(133)
plt.bar(x_poisson, pmf_poisson, color='salmon', alpha=0.7)
plt.title("Poisson Distribution PMF")
plt.xlabel("Number of events")
plt.ylabel("Probability")

plt.tight_layout()
plt.show()


# CONTINUOUS DISTRIBUTIONS

# 1. Normal Distribution (PDF and CDF)
mu, sigma = 0, 1  # mean and standard deviation
x_normal = np.linspace(-5, 5, 1000)
pdf_normal = stats.norm.pdf(x_normal, mu, sigma)
cdf_normal = stats.norm.cdf(x_normal, mu, sigma)

# 2. Uniform Distribution (PDF and CDF)
a, b = 0, 1  # lower and upper bounds
x_uniform = np.linspace(-0.5, 1.5, 1000)
pdf_uniform = stats.uniform.pdf(x_uniform, a, b - a)
cdf_uniform = stats.uniform.cdf(x_uniform, a, b - a)

# 3. Exponential Distribution (PDF and CDF)
lambda_exp = 1  # rate
x_exp = np.linspace(0, 10, 1000)
pdf_exp = stats.expon.pdf(x_exp, scale=1/lambda_exp)
cdf_exp = stats.expon.cdf(x_exp, scale=1/lambda_exp)

# Plotting all distributions
plt.figure(figsize=(14, 10))

# Normal Distribution
plt.subplot(231)
plt.plot(x_normal, pdf_normal, color='blue', label='PDF')
plt.plot(x_normal, cdf_normal, color='red', label='CDF')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.legend()

# Uniform Distribution
plt.subplot(232)
plt.plot(x_uniform, pdf_uniform, color='green', label='PDF')
plt.plot(x_uniform, cdf_uniform, color='purple', label='CDF')
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.legend()

# Exponential Distribution
plt.subplot(233)
plt.plot(x_exp, pdf_exp, color='orange', label='PDF')
plt.plot(x_exp, cdf_exp, color='brown', label='CDF')
plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.legend()

plt.tight_layout()
plt.show()

```

# POPULATION AND SAMPLES & LAW OF LARGE NUMBERS


## Population and Samples

Population: A population refers to the entire set of individuals, items, or data points that share a particular characteristic and are the subject of a statistical study. It includes every possible member of the group you are interested in studying. For example, if you're studying the heights of all 18-year-olds in a country, the population would be every 18-year-old in that country.

Sample: A sample is a subset of the population, selected to represent the population in a statistical study. It is often impractical or impossible to gather data from the entire population, so a sample is chosen, and the results from this sample are used to infer characteristics about the population. The sample must be selected carefully to ensure it is representative, so the conclusions drawn are valid.

### Key differences:

Population refers to the complete set, while a sample is a smaller, manageable subset.
Data from a population is called descriptive statistics, while data from a sample is used for inferential statistics, where conclusions are drawn about the population based on the sample.

### Sampling Methods:

Random Sampling: Every member of the population has an equal chance of being selected. This reduces bias.

Stratified Sampling: The population is divided into subgroups (strata) based on characteristics, and samples are taken from each group.

Systematic Sampling: Every nth individual from the population is selected after choosing a random starting point.

Convenience Sampling: Choosing individuals who are easiest to access or sample, though this can lead to bias.

## Law of Large Numbers

The Law of Large Numbers (LLN) is a fundamental concept in probability and statistics. It states that as the size of a sample increases, the sample mean (or average) will get closer to the population mean (true average). This law provides a foundation for the validity of using sample data to make inferences about the population.

Key idea: The larger the sample size, the more accurate your estimate of the population mean becomes.

### Types of LLN:

Weak Law of Large Numbers: It states that the sample mean will converge in probability to the population mean as the sample size increases. In other words, as you collect more data, the probability that the sample mean will be far from the true mean gets smaller.

Strong Law of Large Numbers: This is a more stringent version. It says that the sample mean will almost surely (with probability 1) converge to the population mean as the sample size goes to infinity.

### Practical interpretation:

If you flip a fair coin, the probability of getting heads (50%) is fixed. If you flip the coin only a few times, you might get heads more or less frequently than 50%, but as the number of flips increases, the average number of heads will get closer and closer to 50%. The more flips you make, the more accurate your estimate of the true probability becomes.

### Implications:

The Law of Large Numbers allows us to trust sample averages when we have large enough samples, which is why larger surveys or experiments yield more reliable results.
This law is why large datasets in machine learning and data science typically give more accurate results compared to smaller ones.

# MEAN, MEDIAN & EXPECTED VALUES

## Mean (Arithmetic Average)
Definition:
The mean is the sum of all values divided by the total number of values. It's used to find the average of a dataset.

Detailed Example:
You and your friends record your weekly coffee expenses:

Friend A: $10

Friend B: $15

Friend C: $20

Friend D: $25

(10 + 15 + 20 + 25) / 4 = 17.5


Interpretation: On average, each person spends $17.50 per week on coffee.

### Limitations:
The mean is sensitive to outliers.

Example: If one person spent $100 on coffee (instead of $25), the mean would jump to:

Mean= (10+15+20+100) / 4 =36.25

This doesn't reflect the majority's spending habits.


## Median
Definition:
The median is the middle value of a dataset when it is ordered in ascending (or descending) order. It divides the dataset into two equal halves—50% of the data lies below it, and 50% lies above it.

### Formula to Calculate Median:
Odd Number of Observations: If the dataset has an odd number of values, the median is the middle value.

Even Number of Observations: If the dataset has an even number of values, the median is the average of the two middle values.

The median is robust to outliers because it depends only on the position of values, not their magnitude.

Example: Incomes of 5 people: $30,000, $35,000, $40,000, $1,000,000 (outlier).

Sorted: $30,000, $35,000, $40,000, $1,000,000.

Median: $37,500 (between $35,000 and $40,000).

Mean would be $276,250, heavily skewed by $1,000,000.


## Expected Value (EV)

The expected value (EV) is the long-term average outcome of a random variable based on its probabilities. It tells us what to expect "on average" over many trials or occurrences.

### Discrete Variables : 

E(X)=  P(x) ⋅ xi

Where:

xi = possible value of the random variable.

P(xi) = probability of xi occurring.
​


```python
# Define the possible outcomes and probabilities
outcomes = [10, 50, 0]  # Amounts won
probabilities = [0.2, 0.05, 0.75]  # Corresponding probabilities

# Cost of the ticket
ticket_cost = 5

# Calculate the net gains (outcomes - cost of ticket)
net_gains = [outcome - ticket_cost for outcome in outcomes]

# Calculate the expected value (EV)
EV_discrete = sum(p * net_gain for p, net_gain in zip(probabilities, net_gains))

print(f"Expected Value (Discrete): ${EV_discrete:.2f}")
```

### Continous Variables

A continuous random variable can take on any value within a given range. Probabilities are not assigned to individual values but instead to intervals, described by a probability density function (PDF).

```python
import numpy as np
from scipy.integrate import quad

# Define the probability density function (PDF) for the exponential distribution
def pdf(x, lambda_):
    return lambda_ * np.exp(-lambda_ * x)

# Define the expected value function (integral of x * PDF(x))
def expected_value_continuous(lambda_):
    # Integrate x * pdf(x) over the range [0, inf]
    EV, _ = quad(lambda x: x * pdf(x, lambda_), 0, np.inf)
    return EV

# Set the lambda parameter for the exponential distribution
lambda_ = 0.1

# Calculate the expected value (EV)
EV_continuous = expected_value_continuous(lambda_)

print(f"Expected Value (Continuous): {EV_continuous:.2f} hours")
```

# VARIANCE, COVARIANCE, & CORRELATION

1. Variance

Definition: Variance measures how much a single variable deviates from its mean. A high variance means the data points are spread out, and a low variance indicates they are close to the mean.<br><br><br>

2. Covariance

Definition: Covariance measures how two variables vary together. If the covariance is positive, the variables increase together. If it's negative, one increases while the other decreases.<br><br><br>

3. Correlation 

Definition: Correlation is a statistical measure that describes how two variables move in relation to each other, on a standardized scale.

Correlation (r) always lies between -1 and 1.

r = 1: Perfect positive correlation (when one variable increases, the other increases proportionally).

r = −1: Perfect negative correlation (when one variable increases, the other decreases proportionally).

r = 0: No linear relationship (the variables are not linearly related, but they may have other relationships).



```python
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [10, 8, 6, 4, 2]

# Variance
var_x = np.var(x, ddof=0)  # Population variance
var_y = np.var(y, ddof=0)  # Population variance
print("Variance of X:", var_x)
print("Variance of Y:", var_y)

# Covariance
cov_matrix = np.cov(x, y, ddof=0)  # Population covariance
covariance = cov_matrix[0, 1]
print("\nCovariance:", covariance)

# Correlation
correlation = np.corrcoef(x, y)[0, 1]
corr_xz = np.corrcoef(x, z)[0, 1]
print("\nCorrelation between X and Y:", correlation)
print("Correlation between X and Z:", corr_xz) 
```

# CENTERAL LIMIT THEOREM & NORMAL DISTRIBUTION

The Central Limit Theorem says that if we take multiple random samples from any population (regardless of the population's original distribution), and calculate the means of those samples, the distribution of those sample means will approach a normal distribution (bell-shaped curve) as the number of samples increases — even if the original population is not normally distributed.

Key Points of CLT:

You can start with any population distribution (uniform, exponential, skewed, etc.).
By sampling from that population and calculating the mean of each sample, the distribution of the sample means will eventually approximate a normal distribution as:

The sample size becomes large enough (typically n > 30).

The number of samples increases. <br><br><br>



A Normal Distribution is a specific type of distribution that is symmetric and bell-shaped. It is characterized by two parameters:

Mean (μ): The average or central value.

Standard Deviation (σ): The spread of values around the mean.

```python
import numpy as np
import matplotlib.pyplot as plt



# Parameters
population_size = 100000
sample_size = 1000
num_samples = 1000

# Generate population data from an exponential distribution
population = np.random.exponential(scale=2, size=population_size)

# Take many random samples and calculate their means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7, label='Sample Means')
plt.title('Central Limit Theorem - Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.legend()
plt.show()





# Parameters for Normal Distribution
mean = 0
std_dev = 1
num_samples = 10000

# Generate random samples from a normal distribution
samples = np.random.normal(mean, std_dev, num_samples)

# Plotting the normal distribution
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot the theoretical normal distribution for comparison
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution (Mean=0, Std Dev=1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

# P-VALUES & CONFIDENCE INTERVALS

## P-VALUES

p-values help us decide whether to reject the null hypothesis (𝐻0) or not. They indicate whether the data provides sufficient evidence to conclude that there is a statistically significant effect or difference. A small p-value suggests that the observed data is unlikely under H0.

Low p-value: Data suggests there is a difference or effect.

High p-value: Data does not show enough evidence to claim a difference or effect (but it doesn’t prove there isn’t one).<br><br>

## CONFIDENCE INTERVALS

Confidence Interval (CI): A confidence interval is a range of values that estimates where the true population parameter (e.g., population mean) is likely to lie. It’s calculated from sample data and has a specific confidence level (like 95% or 99%) that reflects how often the interval would contain the true parameter if the sampling process were repeated many times. A 95% confidence level means that if you were to repeat the sampling process 100 times, about 95 of those calculated confidence intervals would contain the true population mean.  

Calculated Interval: This is the confidence interval computed from your sample. For instance, if you take a sample, calculate the sample mean, and then compute a confidence interval around it, that interval is the "calculated interval." It represents the range in which the true population mean is likely to lie. The confidence interval is calculated around the sample mean by adding and subtracting a margin of error. This margin of error is determined by how spread out the data is (using the standard error, which is based on the standard deviation and sample size) and how large the sample is. Larger samples provide more precise estimates, reducing the margin of error.

(isnt used only for population mean)





```python
# Scenario: Testing a new drug
# Null Hypothesis ( 𝐻 0 : The new drug has no effect (mean improvement is 0).
                 
# Alternative Hypothesis (𝐻𝑎): The new drug has an effect (mean improvement is not 0).

# Data:
# You collect data on 30 patients and find an average improvement of 5 points. Perform a one-sample t-test to test 𝐻 0.



from scipy.stats import ttest_1samp
import numpy as np

# Simulated data: Improvements in patient scores
improvements = np.random.normal(loc=5, scale=2, size=30)  # Mean = 5, Std Dev = 2

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(improvements, 0)  # Test if mean is 0

# Significance level
alpha = 0.05

# Output results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The drug has a significant effect.")
else:
    print("Fail to reject the null hypothesis: No significant effect observed.")


print("\n")


import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from scipy import stats

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Define the model (Support Vector Classifier)
model = SVC()

# Perform 10-fold cross-validation and get accuracy scores
cv_scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')

# Calculate the mean accuracy and the standard error
mean_accuracy = np.mean(cv_scores)
std_error = np.std(cv_scores) / np.sqrt(len(cv_scores))

# Calculate the 95% confidence interval
confidence_level = 0.95

# The Z-score represents how many standard deviations away from the mean we need to go to capture 95% of the possible values (i.e., the range where we expect the true accuracy to lie). 
# For a 95% confidence level, the Z-score is approximately 1.96.
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for 95% confidence

# Confidence interval (lower, upper)
ci_lower = mean_accuracy - z_score * std_error
ci_upper = mean_accuracy + z_score * std_error

# Output the results
print(f"Mean Accuracy: {mean_accuracy:.4f}")
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")

```

# STANDARD DEVIATION, STATISTICAL SIGNIFIGANCE, Z-SCORES


Standard deviation is a measure of how spread out the values in a dataset are around the mean (average). A small SD means the data points are tightly packed around the mean; a large SD means they are widely scattered.<br><br>

Statistical significance refers to whether an observed effect or difference is likely real or if it might have occurred due to random chance.

It is determined by comparing the p-value to a pre-set significance level (α), which is commonly set to 0.05.

If the p-value is less than or equal to the significance level (α), the result is statistically significant, meaning the evidence is strong enough to reject the null hypothesis.

If the p-value is greater than the significance level (α), the result is not statistically significant, meaning there isn’t enough evidence to reject 𝐻0. <br><br>


The z-score (also known as the standard score or z-value) tells you how far a specific data point is from the mean of a dataset, in terms of standard deviations.

Positive Z-score: If 𝑧 > 0, the data point is above the mean.

Negative Z-score: If 𝑧 < 0, the data point is below the mean.

Z-score of 0: If 𝑧 = 0, the data point is exactly at the mean.




```python
import numpy as np
import scipy.stats as stats

# Example dataset
data = [70, 75, 80, 85, 90]

# Calculate the standard deviation
std_dev = np.std(data)
print(f"Standard Deviation: {std_dev}")

print("\n")

# Calculate the z-scores for each data point
z_scores = stats.zscore(data)

# Print the z-scores
print("Z-scores:", z_scores)


```

# HYPOTHESIS TESTING

Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves two competing hypotheses:

Null Hypothesis (H₀): A statement of no effect or no difference.

Alternative Hypothesis (H₁): A statement that contradicts the null hypothesis, suggesting some effect or difference. <br><br>

### Key Concepts:

Significance Level (α): The probability threshold at which you decide whether to reject the null hypothesis, usually set at 0.05 (5%).

p-value: The probability of observing the data, or something more extreme, under the assumption that the null hypothesis is true. If the p-value is less than α, you reject the null hypothesis.

Test Statistic: A standardized value that helps to determine the p-value, which can be calculated based on different statistical tests (e.g., z-test, t-test). <br><br>

### Steps in Hypothesis Testing:

State the Hypotheses: Formulate the null and alternative hypotheses.

Set the Significance Level: Choose α (typically 0.05).

Choose the Appropriate Test: Depending on the data type and assumptions (e.g., z-test, t-test).

Collect Data and Calculate the Test Statistic: Using sample data, calculate the test statistic.

Calculate the p-value: Based on the test statistic.

Make a Decision: Reject or fail to reject the null hypothesis based on the p-value and α.

```python
import numpy as np
from scipy import stats

# Sample data: exam scores of 30 students
sample_data = np.array([70, 80, 85, 72, 78, 74, 65, 88, 90, 85, 
                       76, 83, 79, 75, 72, 71, 80, 82, 77, 68, 
                       74, 79, 81, 73, 76, 70, 78, 80, 81, 72])

# Population mean (from the null hypothesis)
population_mean = 75

# Perform the t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# Output the results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Decision based on p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis (H₀). The population mean is likely not 75.")
else:
    print("Fail to reject the null hypothesis (H₀). The population mean is likely 75.")
```

## TYPES OF HYPOTHESIS TESTS

1. One-Sample t-Test

This test is used when we want to compare the sample mean to a known population mean. We use this when we don't know the population's standard deviation.

```python
import numpy as np
from scipy import stats

# Sample data (exam scores of 30 students)
data = np.array([70, 80, 85, 72, 78, 74, 65, 88, 90, 85, 
                 76, 83, 79, 75, 72, 71, 80, 82, 77, 68, 
                 74, 79, 81, 73, 76, 70, 78, 80, 81, 72])

# Population mean from the null hypothesis (H₀: μ = 75)
population_mean = 75

# Perform the one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, population_mean)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Decision rule
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀). The population mean is not 75.")
else:
    print("Fail to reject the null hypothesis (H₀). The population mean is likely 75.")
```

2. Two-Sample t-Test

This test is used to compare the means of two independent groups. It assumes both samples come from normally distributed populations with unknown but equal variances.

```python
# Data for two groups (group 1 and group 2)
group1 = np.array([50, 60, 55, 70, 65, 75, 85, 90, 92, 80])
group2 = np.array([60, 72, 58, 79, 64, 82, 85, 92, 91, 88])

# Perform the two-sample t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Decision rule
if p_value < alpha:
    print("Reject the null hypothesis (H₀). The means are likely different.")
else:
    print("Fail to reject the null hypothesis (H₀). The means are likely the same.")
```

3. Paired t-Test

This test compares the means of two related groups. It is used when there are two measurements taken from the same subjects or closely related subjects (e.g., before and after treatment).

```python
# Data: Before and after treatment scores for the same group of students
before_treatment = np.array([50, 60, 55, 70, 65, 75, 85, 90, 92, 80])
after_treatment = np.array([60, 72, 58, 79, 64, 82, 85, 92, 91, 88])

# Perform the paired t-test
t_stat, p_value = stats.ttest_rel(before_treatment, after_treatment)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Decision rule
if p_value < alpha:
    print("Reject the null hypothesis (H₀). The means are likely different.")
else:
    print("Fail to reject the null hypothesis (H₀). The means are likely the same.")
```

4. Chi-Square Test

The Chi-Square Test is a statistical method to compare observed data with expected data to determine if there is a significant difference between them. It's most commonly used in two contexts:

Goodness-of-Fit Test: Checks if a sample distribution matches a theoretical distribution.

Test for Independence: Checks if two categorical variables are related or independent.

```python
# Observed frequencies (actual votes for each candidate)
observed = np.array([50, 30, 10, 10])

# Expected frequencies (uniform distribution: 25 votes for each candidate)
expected = np.array([25, 25, 25, 25])

# Perform the chi-square test
chi2_stat, p_value = stats.chisquare(observed, expected)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Decision rule
if p_value < alpha:
    print("Reject the null hypothesis (H₀). The distribution of votes is not as expected.")
else:
    print("Fail to reject the null hypothesis (H₀). The distribution of votes is as expected.")
```

5. ANOVA (Analysis of Variance)

ANOVA is used to compare the means of more than two groups. It tests if at least one group mean is different from the others.

```python
# Data for three groups (teaching methods)
method1 = np.array([50, 60, 55, 70, 65, 75])
method2 = np.array([80, 85, 90, 92, 88, 91])
method3 = np.array([40, 50, 55, 60, 45, 50])

# Perform the one-way ANOVA
f_stat, p_value = stats.f_oneway(method1, method2, method3)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Decision rule
if p_value < alpha:
    print("Reject the null hypothesis (H₀). At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis (H₀). The group means are likely the same.")
```

# SPECIFICITY, SENSITIVITY, & CONFUSION MATRICES


## Confusion Matrix
A confusion matrix is a table that provides a summary of the prediction results on a classification problem. It allows you to calculate various metrics like accuracy, precision, recall (sensitivity), and specificity.

|                | Predicted Positive | Predicted Negative |
|----------------|--------------------|--------------------|
| **Actual Positive** | True Positive (TP)     | False Negative (FN)   |
| **Actual Negative** | False Positive (FP)     | True Negative (TN)     |


True Positive (TP): Model correctly predicts the positive class.

True Negative (TN): Model correctly predicts the negative class.

False Positive (FP): Model incorrectly predicts the positive class (Type I error).

False Negative (FN): Model incorrectly predicts the negative class (Type II error). <br><br>


## Sensitivity (Recall/True Positive Rate)

Sensitivity measures the model’s ability to correctly identify positive instances.

Sensitivity = TP / (TP + FN)

High sensitivity = few false negatives.

Useful when minimizing missed positive cases is crucial, e.g., detecting diseases.<br><br>

## Specificity (True Negative Rate)

Specificity measures the model’s ability to correctly identify negative instances.

Specificity = TN / (TN + FP)

High specificity = few false positives.


Useful when avoiding false alarms is crucial, e.g., screening for cancer in healthy individuals.<br><br>


High Sensitivity ensures most positive cases are detected but might lead to more false positives.

High Specificity ensures most negative cases are correctly ruled out but might lead to more false negatives.

There’s often a tradeoff between the two, managed using a decision threshold. <br><br><br>

Sensitivity and specificity are inversely related for most models. Balancing them often requires adjusting the decision threshold.

Always choose metrics based on the problem’s priorities. For example:

If missing positives is unacceptable → prioritize sensitivity.

If false positives are costly → prioritize specificity.

Use tools like Receiver Operating Characteristic (ROC) curves and Area Under the Curve (AUC) to visualize and balance sensitivity and specificity.


# MULTIPLE COMPARISONS & SOLUTIONS

The multiple comparisons problem arises when a statistical analysis involves multiple simultaneous tests. As the number of tests increases, the likelihood of obtaining at least one statistically significant result due to random chance also increases. This leads to an increased risk of Type I errors (false positives).

Key Points:

Type I Error Rate: In hypothesis testing, a Type I error occurs when a true null hypothesis is incorrectly rejected. The probability of making a Type I error is denoted by the significance level, usually set at 0.05.

Family-Wise Error Rate (FWER): This is the probability of making one or more Type I errors across a family of tests. Without correction, the FWER increases with the number of tests.

Example: If you conduct 20 independent tests at a 5% significance level, the probability of getting at least one false positive is approximately 64% (calculated as (1 - (1 - 0.05)^{20})).<br><br>

## Solutions to the Multiple Comparisons Problem

Several methods have been developed to control the FWER or the False Discovery Rate (FDR) when performing multiple comparisons:

### Bonferroni Correction:

Description: Adjusts the significance level by dividing (\alpha) by the number of tests ((m)). It is simple and widely used.

Pros: Controls FWER effectively.

Cons: Very conservative, especially with a large number of tests, leading to increased Type II errors (false negatives).<br><br>

### Holm-Bonferroni Method:

Description: A step-down procedure that is less conservative than the Bonferroni correction. It sequentially tests hypotheses from the smallest to the largest p-value.

Pros: More powerful than Bonferroni.

Cons: Still controls FWER, may be conservative with many tests.<br><br>

### Benjamini-Hochberg Procedure:

Description: Controls the FDR, which is the expected proportion of false discoveries among the rejected hypotheses.

Pros: More powerful than methods controlling FWER, suitable for large-scale testing (e.g., genomics).

Cons: Allows some false positives, which may not be suitable for all applications.<br><br>

### False Discovery Rate (FDR) Control:

Description: Methods like the Benjamini-Hochberg procedure aim to control the FDR rather than the FWER.

Pros: Balances the rate of Type I and Type II errors, more suitable for exploratory research.

Cons: May not be appropriate when strict control over Type I errors is needed.<br><br>

## Code Example

Corrected p-values: These are the adjusted p-values after applying the respective correction method. They help in determining whether to reject the null hypothesis for each test.

Reject null hypothesis: This boolean array indicates whether each null hypothesis can be rejected at the specified significance level ((\alpha = 0.05)) after applying the correction.


```python
import numpy as np
from statsmodels.stats.multitest import multipletests

# Example p-values from multiple tests
p_values = np.array([0.01, 0.04, 0.03, 0.002, 0.05])

# Apply Bonferroni correction
bonferroni_results = multipletests(p_values, alpha=0.05, method='bonferroni')

print("Bonferroni Corrected p-values:", bonferroni_results[1])
print("Reject null hypothesis (Bonferroni):", bonferroni_results[0])



# Apply Holm-Bonferroni correction
holm_results = multipletests(p_values, alpha=0.05, method='holm')

print("\nHolm-Bonferroni Corrected p-values:", holm_results[1])
print("Reject null hypothesis (Holm-Bonferroni):", holm_results[0])



# Apply Benjamini-Hochberg correction
bh_results = multipletests(p_values, alpha=0.05, method='fdr_bh')

print("\nBenjamini-Hochberg Corrected p-values:", bh_results[1])
print("Reject null hypothesis (Benjamini-Hochberg):", bh_results[0])
```

# CONDITIONAL PROBABILITY & BAYES THEOREM

```python
"""
Understanding Conditional Probability and Bayes' Theorem

1. Conditional Probability:
   P(A|B) = The probability of event A occurring, given that event B has occurred
   Formula: P(A|B) = P(A ∩ B) / P(B)
   where P(A ∩ B) is the probability of both events occurring

2. Bayes' Theorem:
   P(A|B) = P(B|A) * P(A) / P(B)
   where:
   - P(A|B) is the posterior probability
   - P(B|A) is the likelihood
   - P(A) is the prior probability
   - P(B) is the evidence
"""

import numpy as np
from collections import Counter

def conditional_probability_example():
    print("\n1. Basic Conditional Probability Example:")
    print("Scenario: In a deck of cards, what's the probability of drawing a King given that we drew a face card?")
    
    # Total number of cards in a deck
    total_cards = 52
    # Number of face cards (Jack, Queen, King)
    face_cards = 12
    # Number of Kings
    kings = 4
    
    # P(King|Face Card) = P(King ∩ Face Card) / P(Face Card)
    # Since all Kings are face cards, P(King ∩ Face Card) = P(King)
    prob_king_given_face = kings / face_cards
    
    print(f"P(King|Face Card) = {kings}/{face_cards} = {prob_king_given_face}")

def bayes_theorem_example():
    print("\n2. Bayes' Theorem Example:")
    print("Scenario: Medical Test Accuracy")
    print("- Disease prevalence: 1% in population")
    print("- Test sensitivity (true positive rate): 95%")
    print("- Test specificity (true negative rate): 90%")
    
    # Prior probability (disease prevalence)
    P_disease = 0.01
    
    # Likelihood (test sensitivity)
    P_positive_given_disease = 0.95
    
    # False positive rate (1 - specificity)
    P_positive_given_no_disease = 0.10
    
    # Evidence (total probability of positive test)
    P_positive = (P_positive_given_disease * P_disease + 
                 P_positive_given_no_disease * (1 - P_disease))
    
    # Posterior probability using Bayes' theorem
    P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive
    
    print(f"\nIf someone tests positive, probability they have the disease:")
    print(f"P(Disease|Positive) = {P_disease_given_positive:.3f} or {P_disease_given_positive*100:.1f}%")

def practical_example():
    print("\n3. Practical Example with Data:")
    print("Scenario: Email Spam Classification")
    
    # Simulated email data
    emails = [
        ("free money", "spam"),
        ("meeting tomorrow", "ham"),
        ("win prize", "spam"),
        ("project update", "ham"),
        ("lottery winner", "spam"),
        ("lunch meeting", "ham"),
        ("free gift", "spam"),
        ("quarterly report", "ham")
    ]
    
    # Count occurrences
    total_emails = len(emails)
    spam_count = sum(1 for _, label in emails if label == "spam")
    
    # Words that appear in spam
    spam_words = Counter(word for word, label in emails if label == "spam" 
                        for word in word.split())
    
    # Calculate P(spam) - prior probability
    P_spam = spam_count / total_emails
    
    # Example: Calculate P(spam|"free")
    word_to_check = "free"
    
    # P(word|spam) - likelihood
    P_word_given_spam = spam_words[word_to_check] / spam_count if spam_count > 0 else 0
    
    print(f"\nPrior probability of spam: {P_spam:.2f}")
    print(f"Likelihood of '{word_to_check}' in spam: {P_word_given_spam:.2f}")

if __name__ == "__main__":
    print("=== Conditional Probability and Bayes' Theorem Examples ===")
    conditional_probability_example()
    bayes_theorem_example()
    practical_example()
```
