```python
import pandas as pd
```

```python
demographic_data = pd.read_csv('./cleaned_adult_income.csv')
demographic_data.head(15)
```

```python
race_counts = demographic_data["race"].value_counts()
race_counts
```

```python
avg_male_age = demographic_data[demographic_data["gender"] == "Male"]
avg_male_age["age"].mean()
```

```python
percentage_of_bachealors_holders = demographic_data[demographic_data["education"] == "Bachelors"]
percentage_of_bachealors_holders["education"].count() / demographic_data["education"].count() * 100
```

```python
higher_education = demographic_data[demographic_data["education"].isin(["Bachelors", "Masters", "Doctorate"])]

higher_education_percentage = (higher_education["education"].count() / demographic_data["education"].count()) * 100

higher_education_percentage
```

```python
not_higher_education = demographic_data[~demographic_data["education"].isin(["Bachelors", "Masters", "Doctorate"])]

not_higher_education_salary = not_higher_education[not_higher_education["income"] == ">50K"]

not_higher_education_salary_percentage = (not_higher_education_salary["education"].count() / not_higher_education["education"].count()) * 100

not_higher_education_salary_percentage
```

```python
demographic_data.info()
```

```python
min_hours = demographic_data["hours-per-week"].min()
```

```python
minimum_hours_worker = demographic_data[demographic_data["hours-per-week"] == min_hours]

minimum_hours_worker_salary_greater_than_50K = minimum_hours_worker[minimum_hours_worker["income"] == ">50K"]

minimum_hours_worker_salary_greater_than_50K_percentage = (minimum_hours_worker_salary_greater_than_50K["hours-per-week"].count() / minimum_hours_worker["hours-per-week"].count()) * 100

minimum_hours_worker_salary_greater_than_50K_percentage
```

```python
people_make_more_than_50k = demographic_data[demographic_data["income"] == ">50K"]
countries_where_people_make_more_than_50k = people_make_more_than_50k["native-country"].value_counts()

country_with_most_people_making_more_than_50k = countries_where_people_make_more_than_50k.idxmax()
top_country_population = demographic_data[demographic_data["native-country"] == country_with_most_people_making_more_than_50k].shape[0]


print(country_with_most_people_making_more_than_50k)
print((countries_where_people_make_more_than_50k.max() /top_country_population ) * 100)
```
